```python
df_movies = pd.read_csv(
    r'../S05-S08/Data/movies.csv',
    index_col='name',
    parse_dates=['release_date']
)
df_movies
```

```python
above7 = df_movies['score'] >= 7
bellow8 = df_movies['score'] <= 8
df_movies[above7 & bellow8]
```

```python
between7_8 = df_movies['score'].between(7,8)

df_movies[between7_8]
```

```python
release_2010_2016 = df_movies['release_date'].between('2010-01-01','2016-01-01')

df_movies[release_2010_2016]
```

```python
df_movies['genre'].unique()
```

```python
is_mystery = df_movies['genre'] == 'Mystery'
is_Sci_Fi = df_movies['genre'] == 'Sci-Fi'
df_movies[is_Sci_Fi | is_mystery].head(10)
```

```python
mask = df_movies['genre'].isin(['Mystery','Sci-Fi'])
df_movies[mask].head(10)
```
